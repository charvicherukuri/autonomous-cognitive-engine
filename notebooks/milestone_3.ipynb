```python
# Import required modules
import sys
import os
sys.path.append('../src')

from agents.supervisor_agent import SupervisorAgent
from tools.delegation_tool import analyze_task_for_delegation, get_available_agents
```

## Test 1: Task Analysis and Delegation

```python
# Test task analysis
test_tasks = [
    "Create a strategic plan for launching a new product",
    "Write Python code for data analysis",
    "Write a blog post about AI trends",
    "Research the latest developments in quantum computing",
    "Summarize this quarterly business report",
    "Translate this document to Spanish"
]

print("Task Analysis Results:")
print("=" * 50)

for task in test_tasks:
    analysis = analyze_task_for_delegation(task)
    print(f"\nTask: {task[:40]}...")
    print(f"Recommended Agent: {analysis['recommended_agent']}")
    print(f"Confidence: {analysis['confidence']:.2f}")
    print(f"Reasoning: {analysis['reasoning']}")
```

## Test 2: End-to-End Delegation

```python
# Test complete delegation workflow
supervisor = SupervisorAgent()

delegation_tests = [
    "Create a project timeline for developing a mobile app",
    "Write a Python function to calculate fibonacci numbers",
    "Write marketing copy for a new AI product launch"
]

print("End-to-End Delegation Tests:")
print("=" * 50)

for i, test in enumerate(delegation_tests, 1):
    print(f"\nTest {i}: {test}")
    print("-" * 40)
    
    result = supervisor.run(test)
    
    print(f"Success: {result['success']}")
    print(f"Delegated to: {result.get('delegated_to', 'None')}")
    print(f"Tools used: {result.get('tools_used', 0)}")
    print(f"Response preview: {result['final_response'][:200]}...")
```

## Test 3: Available Agents

```python
# Display available agents
agents = get_available_agents()

print("Available Sub-Agents:")
print("=" * 50)

for agent in agents:
    print(f"\nAgent: {agent['name']}")
    print(f"Type: {agent['type']}")
    print(f"Capabilities: {', '.join(agent['capabilities'][:3])}...")
    print(f"Optimal for: {', '.join(agent['optimal_for'][:2])}...")
```
