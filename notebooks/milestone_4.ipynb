## Test 1: Translation Agent Capabilities

```python
# Test Translation Agent
translation_agent = TranslationAgent()

# Get capabilities
capabilities = translation_agent.get_capabilities()
print("Translation Agent Capabilities:")
print("=" * 50)
print(f"Agent Type: {capabilities['agent_type']}")
print(f"Name: {capabilities['name']}")
print(f"Description: {capabilities['description']}")
print(f"Supported Languages: {capabilities['language_count']}")
print(f"Ready: {capabilities['ready']}")

print("\nCapabilities:")
for capability in capabilities['capabilities']:
    print(f"  - {capability}")

print("\nOptimal for:")
for use_case in capabilities['optimal_for']:
    print(f"  - {use_case}")
```

## Test 2: Translation Functionality

```python
# Test basic translation
test_text = "Welcome to our enterprise AI platform. We provide cutting-edge solutions for business automation."

translation_tests = [
    ("spanish", "Spanish"),
    ("french", "French"),
    ("german", "German"),
    ("japanese", "Japanese")
]

print("Translation Tests:")
print("=" * 50)
print(f"Original Text: {test_text}")
print()

for lang_code, lang_name in translation_tests:
    result = translation_agent.translate_text(test_text, lang_code)
    
    if result['success']:
        print(f"{lang_name}: {result['translation']}")
    else:
        print(f"{lang_name}: Translation failed - {result['error']}")
    print()
```

## Test 3: Multilingual Content Creation

```python
# Test multilingual content creation
base_content = "Thank you for choosing our premium service. We're committed to your success."
target_languages = ["spanish", "french", "german"]

result = translation_agent.create_multilingual_content(
    base_content=base_content,
    target_languages=target_languages,
    content_type="business"
)

print("Multilingual Content Creation:")
print("=" * 50)
print(f"Base Content: {base_content}")
print()

if result['success']:
    print(f"Successfully created content in {result['successful_translations']} languages:")
    print()
    
    for language, content in result['multilingual_content'].items():
        print(f"{language.title()}: {content}")
        print()
else:
    print(f"Failed to create multilingual content: {result['error']}")
```

## Test 4: End-to-End Translation via Supervisor

```python
# Test translation through supervisor agent
supervisor = SupervisorAgent()

translation_requests = [
    "Translate this to Spanish: Hello world, how are you today?",
    "Create multilingual versions in French and German: Welcome to our platform",
    "Translate this business message to Japanese: We appreciate your partnership"
]

print("End-to-End Translation Tests:")
print("=" * 50)

for i, request in enumerate(translation_requests, 1):
    print(f"\nTest {i}: {request}")
    print("-" * 40)
    
    result = supervisor.run(request)
    
    print(f"Success: {result['success']}")
    print(f"Delegated to: {result.get('delegated_to', 'None')}")
    print(f"Response: {result['final_response'][:200]}...")
```
